## 题目：寻找关键用户和分析社交关系

### data.json 数据描述
#### 1. 用户节点数据包含用户ID和他们的"影响力分数"，格式为["用户名称", 影响力分数]。
#### 2. 边关系数据包含每两个用户之间的连接强度（即边的权重），格式为["用户名称1", "用户名称2", 关系权重]。
例如：
- 节点：["Yanyan-Wang", 50.95], ["Aarebecca", 23.82], ...
- 边：["Yanyan-Wang", "Aarebecca", 25.3], ["Yanyan-Wang", "pomelo-nwu", 25.09], ...

### 问题
#### 1. 社交网络构建与分析
- a. 使用邻接矩阵或者字典表示法来表示这个社交关系网络。
- b. 请找出每个用户的直接朋友（即与其存在边关系的节点），并用列表形式列出。

#### 2. 统计与度中心性分析
- a. 计算每个用户的度数（即直接朋友数量），列出度数最多的前5个用户及其朋友数量。
- b. 通过影响力分数（如上所示），找出影响力最高的前5个用户。
- c. 综合考虑影响力分数和朋友数量，列出对社交网络最有影响力的用户（可以选择将两者按一定比例加权，例如影响力得分的50%和朋友数量的50%）。

#### 3. 推荐新朋友
- a. 对于用户 Yanyan-Wang，请找出其“朋友的朋友”中最有可能成为新朋友的用户，并列出推荐的理由（例如有最多共同的朋友）。
- b. 考虑用户之间的连接强度，找出与 Yanyan-Wang 连接强度最高的5个用户（无论是否已经是直接朋友），并列出推荐排序。
 
#### 4. 网络图的可视化
- a. 使用图论可视化工具绘制用户关系网络图，展示关键用户和他们的连接情况。
- b. 你可以对图中的节点按“影响力得分”进行着色，节点的大小则代表度数，尝试解释在图中哪些用户在社交网络中最具影响力。

```python
import numpy as np
import pandas as pd
import json
import networkx as nx
import matplotlib.pyplot as plt

# 读取数据
with open('hw4_data.json', 'r', encoding='utf-8') as f:
    data = json.load(f)

# 创建社交得分、社交关系列表
names_list=[user[0] for user in data['nodes']]
scores_list=pd.DataFrame(data['nodes'],columns=['name','score'])
social_network=pd.DataFrame(data['edges'],columns=['name1','name2','scores'])
# 创建社交关系矩阵
social_network_matrix=np.zeros((len(names_list),len(names_list)))
social_network_matrix=pd.DataFrame(social_network_matrix,index=names_list,columns=names_list)

#社交关系矩阵的元素，设置为社交得分
for i in range(len(social_network)):
    social_network_matrix.loc[social_network.iloc[i,0],social_network.iloc[i,1]]=social_network.iloc[i,2]
    social_network_matrix.loc[social_network.iloc[i,1],social_network.iloc[i,0]]=social_network.iloc[i,2]

# 保存社交关系矩阵为txt、csv文件
# with open('social_network_matrix.txt', 'w', encoding='utf-8') as f:
#     f.write(str(social_network_matrix))
# social_network_matrix.to_csv('social_network_matrix.csv', encoding='utf-8',index=True,float_format='%.2f')

# 创建字典，用于存储每个用户的朋友数量，最终合并到scores_list中
friend_num = social_network_matrix.gt(0).sum(axis=1).to_frame(name='friend_num')
scores_list=pd.merge(scores_list,friend_num,left_on='name',right_index=True)

# 找出每个用户的直接朋友并用列表表示出来
direct_friends_list=[]
for i in range(len(names_list)):
    direct_friends=social_network_matrix.loc[names_list[i],social_network_matrix.loc[names_list[i]]!=0].index.tolist()
    direct_friends_list.append(direct_friends)
direct_friends_list=pd.DataFrame(direct_friends_list,index=names_list)

#排名
friend_num.sort_values(by='friend_num',ascending=False,inplace=True)
for i in range(0,5):
    print(f"好友数排名第{i+1}的朋友数为:{friend_num.iloc[i,0]}，对应的用户为：{friend_num.index[i]}")
scores_list.sort_values(by='score',ascending=False,inplace=True)
print("\n")
for i in range(0,5):
    print(f"影响力排名第{i+1}的用户的影响力为:{scores_list.iloc[i,1]}，对应的用户为：{scores_list.iloc[i,0]}")
fin_scores=scores_list['score']*0.5+scores_list['friend_num']*0.5
scores_list['final_score']=fin_scores
scores_list.sort_values(by='final_score',ascending=False,inplace=True)
print("\n")
for i in range(0,5):
    print(f"综合排名第{i+1}的用户的综合得分为:{scores_list.iloc[i,3]}，对应的用户为：{scores_list.iloc[i,0]}")

# 推荐Yanyan-Wang的新朋友（共同好友）
yanyan_friends = direct_friends_list.loc['Yanyan-Wang'].tolist()
potential_friends = {}
friends_of_friend = direct_friends_list.loc['Yanyan-Wang'].tolist()
for friend in yanyan_friends:
     if friend is not None: 
        friends_of_friend = direct_friends_list.loc[friend].tolist()
        for fof in friends_of_friend:
            if fof is not None:
                if fof not in potential_friends:
                    potential_friends[fof] = 1
                else:
                    potential_friends[fof] += 1
# 推荐Yanyan-Wang的新朋友（共同好友）
print("\n")
potential_friends_list = pd.DataFrame(list(potential_friends.items()),columns=['name','score'])
potential_friends_list.sort_values(by='score',ascending=False,inplace=True)
print("推荐Yanyan-Wang的新朋友为：")
for i in range(0,5):
    print(f"第{i+1}位的用户为：{potential_friends_list.iloc[i,0]}，共同好友数量为：{potential_friends_list.iloc[i,1]}")

# 绘制社交网络图
# 创建图形对象
G = nx.Graph()
# 添加节点
G.add_nodes_from(names_list)
# 添加加权边
edges = social_network.values
if edges.size > 0:  # 确保社交网络数据不为空
    G.add_weighted_edges_from([(edge[0], edge[1], edge[2]) for edge in edges])
# 计算节点位置
pos = nx.spring_layout(G, k=0.15, iterations=20)  # 调整弹簧常数以获得更好的布局
# 绘制节点、边和标签
plt.figure(figsize=(12, 8))  # 增大图形窗口
nx.draw_networkx_nodes(G, pos, node_size=300, node_color='lightblue', alpha=0.8)  # 调整节点大小和颜色
nx.draw_networkx_edges(G, pos, width=2, alpha=0.5, edge_color='lightgray')  # 调整边宽度和颜色
nx.draw_networkx_labels(G, pos, font_size=12, font_family='sans-serif', font_color='black')
# 设置图形参数
plt.title("社交网络图", fontsize=15)  # 添加标题
plt.axis('off')
plt.tight_layout()  # 适应图形大小
plt.show()
scores_list.reset_index(drop=True, inplace=True)  # 重置索引

# 图中的节点按“影响力得分”进行着色，节点的大小则代表度数
node_sizes = scores_list['score'] * 100  # 节点大小
node_colors = scores_list['score']  # 节点颜色
sc = nx.draw_networkx_nodes(G, pos, node_size=node_sizes, node_color=node_colors, cmap=plt.cm.Reds, alpha=0.8)

nx.draw_networkx_edges(G, pos, width=2, alpha=0.5, edge_color='lightgray')
nx.draw_networkx_labels(G, pos, font_size=12, font_family='sans-serif', font_color='black')

plt.title("社交网络图", fontsize=15)
plt.axis('off')

# 添加颜色条，使用创建的 scatter 对象
plt.colorbar(sc, label='影响力得分')  # 添加颜色条
plt.tight_layout()
plt.show()
```
